# Подключение зависимостей

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tag import pos_tag
from nltk.tokenize import word_tokenize
from nltk.util import ngrams, bigrams

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

# Считывание содержимого файла в строку

```python
s = ''
stop_words = ''
with open('/content/sample_data/hp.txt', encoding='cp1252') as f:
  s = f.read()

with open('/content/sample_data/stops.txt', encoding='cp1252') as f:
  stop_words = f.read().split()

```

# Приведение текста к стандарту


```python
# Приведение текста к нижнему регистру
s = s.lower()

# Удаление чисел, спецсимволов, знаков препинания и url-ссылок
s = re.sub(r'[^\w\s]', ' ', s)
s = re.sub(r'\d+', ' ', s)
s = re.sub(r'http\S+', ' ', s)

# Удаление стоп-слов
stop_words = stop_words + stopwords.words('english')
word_tokens = word_tokenize(s)
s = [word for word in word_tokens if word not in stop_words]

s = ' '.join(s)
# Лемматизация
def lemmatize_all(sentence):
    lemmatizer = WordNetLemmatizer()
    for word, tag in pos_tag(word_tokenize(sentence)):
        if tag.startswith("NN"):
            yield lemmatizer.lemmatize(word, pos='n')
        elif tag.startswith('VB'):
            yield lemmatizer.lemmatize(word, pos='v')
        elif tag.startswith('JJ'):
            yield lemmatizer.lemmatize(word, pos='a')
        else:
            yield word

lems = lemmatize_all(s)
s = ' '.join(lems).split()
print(s)
```

# Мешок слов

```python
bag_of_words = {}
for word in s:
  if word not in bag_of_words:
    bag_of_words[word] = 1
  else:
    bag_of_words[word] += 1


bag_of_words = dict(sorted(bag_of_words.items(), key=lambda item: item[1], reverse=True))
print(len(bag_of_words))
print(bag_of_words)
```

# N-граммы

```python
# 2-граммы
two_grams = list(bigrams(s))

print(len(two_grams), two_grams)

# 4-граммы
four_grams = list(ngrams(s, 4))
print(len(four_grams), four_grams)
```
